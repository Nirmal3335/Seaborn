```python
#Bar plot
```

```python
student = {"nanthini":98,"nirmal":80,"puvi":35,"kavi":65}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.title("marks of student")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
#horizontal bar-plot "barh"
```

```python
plt.barh(names,values,color="orange")
plt.title("marks of student")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,5,7,9,3,1,6,4,6]
b = [5,2,6,8,1,3,4,1,7]

plt.scatter(x,a,marker="*",c="green",s=200)
plt.scatter(x,b,c="blue",s=500)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="green",s=200)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="red",s=600)

plt.show()
```

```python
#pie-chart
```

```python
fruit = ["apple","orange","guava","mango","graph"]
quantity =[52,85,62,74,31]

plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct="%0.1f%%",colors=["green","blue","pink","red","gray"])
plt.show()
```

```python
#doughNut chart
```

```python
plt.pie(quantity,labels=fruit,autopct="%0.1f%%",colors=["green","blue","pink","red","gray"],radius=2)
plt.pie([2],colors="w",radius=1)
plt.show()
```
